```python
import folium

# Criar um objeto de mapa
mapa = folium.Map(location=[-22.767823911381047,-46.26920747744522], zoom_start=12)

# Adicionar um marcador
icone=folium.Icon(color='red')
folium.Marker(location=[-22.767823911381047,-46.26920747744522], icon=icone, popup='Centro').add_to(mapa)
folium.Marker(location=[-23.150430, -45.778035], popup='Grupo 1').add_to(mapa)
folium.Marker(location=[-22.731910, -47.454605], popup='Grupo 2').add_to(mapa)
folium.Marker(location=[-23.558516, -46.713095], popup='Grupo 3').add_to(mapa)
folium.Marker(location=[-21.170440, -47.785087], popup='Grupo 4').add_to(mapa)
folium.Marker(location=[-22.314190, -49.062805], popup='Grupo 5').add_to(mapa)
folium.Marker(location=[-23.061994, -47.167170], popup='Grupo 6').add_to(mapa)
folium.Marker(location=[-22.035607, -47.876790], popup='Grupo 7').add_to(mapa)

folium.LayerControl().add_to(mapa)
# Mostrar o mapa
mapa
```
